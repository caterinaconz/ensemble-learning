```python

df_style = pd.read_csv('/content/drive/MyDrive/Ensemble Learning/Tweets/style_data.csv',
                             index_col=0)

df_bow = pd.read_csv("/content/drive/MyDrive/Ensemble Learning/Tweets/bag of words.csv",
                             index_col=0)
# assert that dataframes are of same length and the the indexes are the same
assert len(df_style) == len(df_bow) 
assert (df_style.index != df_bow.index).any()==False
```

```python

del_topic = ['tokens', 'tokens_back_to_text', 'lemmas', 'lemmas_back_to_text', 'lemma_tokens']
df_bow.drop(del_topic,axis=1, inplace=True)
```

```python
df_all_features = pd.concat([df_style,df_bow],axis=1)
# remove duplicated columns (agency)
df_all_features = df_all_features.loc[:, ~df_all_features.columns.duplicated()] 
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer


# Write code to convert the list of documents to list of tokens.


# Write code to create a TfidfVectorizer object
tfidf = TfidfVectorizer(max_features = 1000)


# Write code to vectorize the sample text
X_tfidf = tfidf.fit_transform(df_all_features['tweet_text'])
X_tfidf_df = pd.DataFrame(X_tfidf.toarray())
df = pd.concat([X_tfidf_df,df_all_features], axis = 1)
df.head()
```
